```python
import requests
import re

GROQ_API_KEY = "insert-api-key-here"
MAX_CHARS = 12000  # ~3000 tokens

def trim_to_sentences(text, max_chars):
    # Split into sentences while preserving punctuation
    sentences = re.findall(r'.*?[.!?](?:\s|$)', text, flags=re.DOTALL)

    if not sentences:  # No punctuation-based sentences
        if len(text) <= max_chars:
            return text
        else:
            raise ValueError("Input too long and contains no sentence boundaries to trim safely.")

    trimmed = ""
    for sentence in sentences:
        if len(trimmed) + len(sentence) <= max_chars:
            trimmed += sentence
        else:
            # If even the first sentence is too long → error
            if trimmed == "":
                raise ValueError(f"Single sentence exceeds max limit ({max_chars} chars): {sentence[:100]}...")
            break

    return trimmed

def machine_humanize(text):
    text = trim_to_sentences(text, MAX_CHARS)

    url = "https://api.groq.com/openai/v1/chat/completions"
    headers = {
        "Authorization": f"Bearer {GROQ_API_KEY}",
        "Content-Type": "application/json"
    }

    data = {
        "model": "llama-3.3-70b-versatile",
        "messages": [
            {"role": "system", "content": "Rewrite the following text so it sounds naturally human-written. Output ONLY the rewritten text."},
            {"role": "user", "content": text}
        ],
        "temperature": 0.7
    }

    response = requests.post(url, json=data, headers=headers)
    response.raise_for_status()
    return response.json()["choices"][0]["message"]["content"].strip()
```
